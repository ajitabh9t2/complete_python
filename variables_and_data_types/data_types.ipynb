```python
# type of operators:


#1. Arithmetic operators: +, -, *, /, //, %, ** , e.g. a+b, a-b, a*b, a/b, a//b, a%b, a**b
a=2
b=7
sum=a+b
print("The sum of", a, "and", b, "is","e.g a+b=", sum)

a=10
b=3
print ("a/b =", a/b)   # division

print ("a//b =", a//b) # floor division means it gives the quotient without the remainder

print ("a%b =", a%b)   # modulus gives the remainder

print ("a**b =", a**b) # exponentiation means a raised to the power of b



#2. Comparison operators: ==, !=, >, <, >=, <= , e.g. a==b, a!=b, a>b, a<b, a>=b, a<=b


#3. Logical operators: and, or, not, e.g. a>0 and b>0, a>0 or b>0, not(a>0)

print("Logical operators:")
a = 5
b = 10
print("a > 0 and b > 0:", a > 0 and b > 0, "both a and b are greater than 0")
print("a > 0 or b > 0:", a > 0 or b > 0, "at least one of a or b is greater than 0")
print("not(a > 0):", not(a > 0), "a is not greater than 0")



#4. Assignment operators: =, +=, -=, *=, /=, //=, %=, **= , e.g. a=5, a+=2, a-=2, a*=2, a/=2, a//=2, a%=2, a**=2

num=10
print("Initial value of num:", num)
num += 5
print("After num += 5:", num)
num -= 3
print("After num -= 3:", num)
num *= 2
print("After num *= 2:", num)
num /= 4
print("After num /= 4:", num)
num //= 2
print("After num //= 2:", num)
num %= 3
print("After num %= 3:", num)
num **= 3
print("After num **= 3:", num)


#5. Bitwise operators: &, |, ^, ~, <<, >>, e.g. a&b, a|b, a^b, ~a, a<<1, a>>1
a = 5  # in binary: 0101
b = 3  # in binary: 0011   

print("Bitwise operators:")
print("a & b:", a & b, "bitwise AND of a and b")
print("a | b:", a | b, "bitwise OR of a and b")
print("a ^ b:", a ^ b, "bitwise XOR of a and b")
print("~a:", ~a, "bitwise NOT of a")
print("a << 1:", a << 1, "left shift a by 1 bit")
print("a >> 1:", a >> 1, "right shift a by 1 bit")



#6. Identity operators: is, is not, e.g. a is b, a is not b
x = [1, 2, 3]
y = [1, 2, 3]
print("Identity operators:")
print("x is y:", x is y, "x and y are different objects in memory")
print("x is not y:", x is not y, "x and y are different objects in memory")
    


#7. Membership operators: in, not in, e.g. 'a' in 'apple', 'b' not in 'apple'
print("Membership operators:")
print("'a' in 'apple':", 'a' in 'apple', "'a' is in 'apple'")
print("'b' not in 'apple':", 'b' not in 'apple', "'b' is not in 'apple'")



#relational operators are also called comparison operators, they compare the values of two operands and return a boolean result (True or False) based on the comparison. The relational operators in Python include:
#1. == (equal to): Returns True if the values of the operands are equal, otherwise returns False.
a = 5
b = 10
print("a == b:", a == b, "a is equal to b")


#2. != (not equal to): Returns True if the values of the operands are not equal, otherwise returns False.
print("a != b:", a != b, "a is not equal to b")

#3. > (greater than): Returns True if the value of the left operand is greater than the value of the right operand, otherwise returns False.
print("a > b:", a > b, "a is greater than b")


#4. < (less than): Returns True if the value of the left operand is less than the value of the right operand, otherwise returns False.
print("a < b:", a < b, "a is less than b")


#5. >= (greater than or equal to): Returns True if the value of the left operand is greater than or equal to the value of the right operand, otherwise returns False.
print("a >= b:", a >= b, "a is greater than or equal to b")


#6. <= (less than or equal to): Returns True if the value of the left operand is less than or equal to the value of the right operand, otherwise returns False.   
print("a <= b:", a <= b, "a is less than or equal to b") 

```

```python
#slicing of strings: means extracting a portion of the string using the syntax string[start:end],
#  where start is the index of the first character to include and end is the index of the first character to exclude. 
# For example, my_string[0:5] would return "Hello", and my_string[6:11] would return "World".

#str [starting_index:ending_index]
str= "ApnaCollege"
print(str[0:4])  # Output: "Apna"
print(str[4:11]) # Output: "College"
print(str[4:len(str)]) # Output: "College"
print(str[4:])
print(str[:]) 


```

```python
#negative indexing in strings: 

# means counting from the end of the string instead of the beginning. The last character of the string has an index of -1, the second to last character has an index of -2, and so on.
# For example, my_string[-1] would return "!", and my_string[-6:-1] would return "World".
str="Apple"
print(str[-1])  # Output: "e"
print(str[-6:-1])  # Output: "Appl"
```

```python
#WAP for a complete calculator work when it is asked to perform a specific operation (addition, subtraction, multiplication, division) on two numbers input by the user. 
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))    
operation = input("Enter the operation (+, -, *, /): ")
if operation == "+":
    result = num1 + num2
    print("Result:", result)
elif operation == "-":
    result = num1 - num2
    print("Result:", result)
elif operation == "*":
    result = num1 * num2
    print("Result:", result)
elif operation == "/":
    if num2 != 0:
        result = num1 / num2
        print("Result:", result)
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Invalid operation. Please enter one of +, -, *, /.")

    #in the above code make itthat after performing an operation again ask the user if they want to perform another operation, and if they say yes, repeat the process, otherwise exit the program.
while True:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))    
    operation = input("Enter the operation (+, -, *, /): ")
    if operation == "+":
        result = num1 + num2
        print("Result:", result)
    elif operation == "-":
        result = num1 - num2
        print("Result:", result)
    elif operation == "*":
        result = num1 * num2
        print("Result:", result)
    elif operation == "/":
        if num2 != 0:
            result = num1 / num2
            print("Result:", result)
        else:
            print("Error: Division by zero is not allowed.")
    else:
        print("Invalid operation. Please enter one of +, -, *, /.")
    
    again = input("Do you want to perform another operation? (yes/no): ")
    
    # whay we used again here is to ask the user if they want to perform another operation after the first one is completed. 
    #If the user enters "yes", the loop will continue and they can perform another operation. If they enter anything other than "yes", the loop will break and the program will exit. 


    if again.lower() != "yes":                 # if the user does not enter "yes", the loop will break and the program will exit.
         


         # The .lower() method is used to convert the user's input to lowercase, so that it can handle inputs like "Yes", "YES", or "yEs" as well.

         print("Exiting the calculator. Goodbye!")

        #want to use break statement to exit the loop when the user does not want to perform another operation.
        
         break

```

```python
#how to use break statement in python:
# The break statement in Python is used to exit a loop prematurely when a certain condition is met. It can be used in both for and while loops. When the break statement is executed, the loop is immediately terminated, and the program continues with the next statement after the loop. 
# Here is an example of using the break statement in a while loop:
while True:
    user_input = input("Enter a number (or 'exit' to quit): ")
    if user_input.lower() == 'exit':
        print("Exiting the loop. Goodbye!")
        break  # This will exit the loop when the user types 'exit'
    else:
        try:
            number = float(user_input)
            print("You entered:", number)
        except ValueError:
            print("Invalid input. Please enter a number or 'exit'.")
```
